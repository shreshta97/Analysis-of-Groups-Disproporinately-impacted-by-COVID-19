# Analysis of Groups Disproporinately impacted by COVID-19

## Background
#### Many of us face the daily reality that SARS CoV-2, a recently identified human coronavirus, has caused a worldwide pandemic. This respiratory illness is spread through droplet-based transmission and leads to severe respiratory problems, kidney failure, and even death. This disease affects almost everyone's daily lives but may be said to impact certain groups disproportionately. It is commonly known that people with certain medical conditions and specific age categories are at an increased risk for severe disease following exposure to COVID-19. However, the impact of race, geographic location, economic status, etc. is not often acknowledged. According to the CDC, " Long-standing systemic health and social inequities have put many people from racial and ethnic minority groups at increased risk of getting sick and dying from COVID-19"(CDC.gov). Some of the factors that may contribute to this increased risk could include racial discrimination, healthcare access, educational gaps, income gaps, and wealth gaps. This is an instance that brings forward the idea of health care disparities, which are differences among health and health care between various population groups. For example, "In the U.S. African Americans are contracting SARS-CoV-2 at higher rated and are more likely to die from COVID-19"(Del Rio 2020).

## Why is this issue important?
#### Health disparities adversely impact groups of people who have, throughout history, experienced more significant obstacles to health based on race, religion, socioeconomic status, gender, age, mental health, or disability. These disparities do not solely impact the group facing differences. Still, they can also limit overall gains in quality of care from the broader population due to the burden of unnecessary cost(health care cost and lost productivity costs). Specifically, in the case of COVID-19, addressing these disparities will allow for better contact tracing and containment of the disease, which is beneficial to all society members. It is crucial to recognize precisely where they stem from and who is being impacted at the most astounding rate to address these disparities. So we should turn to the data to examine where the disease hot spots are and why they are more severely affected. Can we find any population trends that are common in these locations? With this knowledge, public health officials become better equipped to address the public's needs through education, targeted health programs, and more advanced surveillance efforts.

```
#Import libaries needed to clean data 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```
#Import the States csv as a pandas dataframe and define the columns
# This data set was retrived from git hub https://github.com/nytimes/covid-19-data
df_Covid = pd.read_csv('us-states.csv', header=None)
df_Covid.columns = ["date","state","fips","cases","deaths"]
```

# Variable definitions: 
- date: starting January 21st, 2020 to October 21st 2020 when it was downloaded
- state: included in the united states and its surrounding territories
- fips: Federal Infomation processing standard codes that identify unique geographic regions
- cases: The total number of cases of Covid-19, including both confirmed and probable
- deaths: The total number of deaths from Covid-19, including both confirmed and probable.

```
# find the length of the data frame to see the scale of data that we are working with
len(df_Covid)
```

```
# preview the data fram and rename it df1 for ease in future manipulation
df1 = df_Covid[1:]
df1.head()
```

```
# convert each value in the cases and deaths column to integers
df1['cases'] = df1['cases'].astype(int)
df1['deaths']= df1['deaths'].astype(int)
```

```
# create a dataframe from df1 which sums the deaths and cases in each state. 
sum_df= df1.groupby('state')["cases","deaths"].agg(['sum'])
sum_df
```

# Analysis of COVID-19 Cases v.s. Deaths: 
##### As expected, there is a positive correlation between coronavirus cases and deaths seen in Fig 1.1

```
# plot sum of cases v.s. sum of deaths in order to analyze the seemingly obvious correlation between coronavirus cases and deaths for this reason we will be using case load for all futher anylysis
sum_df.plot(kind='scatter',x="cases",y="deaths")
plt.suptitle('COVID-19 Deaths V.S. Cases Fig 1.1')
```

# Analysis of Poverty Indicators in Relation to COVID-19:

```
# import poverty indicator data retrieved from US census find the length to see how big the data set is
df_Income = pd.read_csv('states_poverty.csv', header=None)
len(df_Income)
```

```
# make a new data frame to remove the top few rows and clean the data a bit
df3 = df_Income[3 :]
df3.head()
```

# 5 states with highest Median Household Income: 
1. Tennessee
2. Michigan
3. North Carolina

5. Illinois

```
# create a smaller data frame from the poverty indicators data set to only include median household income
small_df3 = df3[[2, 21]]
small_df3.head()
```

```
# deine the column names and index of the new data frame
small_df3.drop(small_df3.index[1])
small_df3.columns = ["Name","Median Household Income"]
small_df3
```

```
# sort the data frame by increasing household income
small_df3.sort_values(by=['Median Household Income','Name'])
```

```
# rename the colunm Name to be state for easing later merges
small_df4= small_df3.rename(columns = {'Name': 'state'}, inplace = False)
```

```
# use the merge function to create a combination data frame that includes state, cases, death and median household income
combo_df= pd.merge(df2, small_df4)
combo_df
```

```
# sort the merged data frame in increasing order by cases
combo_df= combo_df.sort_values(by=['cases','state'])
```

# Corelation between Coronavirus Cases/Deaths and Median Household Income:
### - There appears to be a slight positive correlation for both Coronavirus cases and deaths but not by much as can be seen in Fig 2.1-2.2

```
# the commas in the data frame need to be removed to process the Median Household Income as an integer
combo_df = combo_df.replace(',','', regex=True)
```

```
# convert the Median Household Income colum to integers
combo_df['Median Household Income'] = combo_df['Median Household Income'].astype(int)
```

```
# plot the Cases vs income and Deaths vs income to see if any correlations exist
combo_df.plot(kind='scatter',x= 'Median Household Income',y=["cases"])
plt.xlabel("Median Household Income")
plt.ylabel("COVID-19 Cases")
plt.suptitle('Cases V.S. Income Fig 2.1')

combo_df.plot(kind='scatter',x="Median Household Income", y="deaths")
plt.xlabel("Median Household Income")
plt.ylabel("COVID-19 Deaths")
plt.suptitle('Deaths V.S. Income Fig 2.2')
```
